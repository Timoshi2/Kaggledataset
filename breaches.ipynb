This is a kaggle dataset on the world's biggest breaches from 2004-2025. Only breaches exposing 30,000+ records are included, establishing a minimum severity threshold. Each record represents a single breach incident with details on the affected organization, breach method, data sensitivity, and records exposed. The primary sources are Identity Theft Resource Center (ITRC), DataBreaches.net, HHS Breach Portal, and News outlets (NYT, BBC, Guardian, Wired, Tech Crunch, Forbes, etc.).This dataset is a good source to use because of the many sources it has backing it up and the recency it contains as it was updated 3 months ago.

```python
sns.set_style('whitegrid')
sns.lmplot(x='year', y='records lost', data=df,
           palette='coolwarm', height=6, aspect=1, fit_reg=False)
```

Looking at this graph showing records lost and the corresponding years, most of the records lost were in the hundred thousands while in the year 2024 and 2023 it seems to reach its peak in records lost reaching the hundreds of millions. 2025 seems to be the tamest of all the years.
