This is a kaggle dataset on the world's biggest breaches from 2004-2025. Only breaches exposing 30,000+ records are included, establishing a minimum severity threshold. The primary sources are Identity Theft Resource Center (ITRC), DataBreaches.net, HHS Breach Portal, and News outlets (NYT, BBC, Guardian, Wired, Tech Crunch, Forbes, etc.).This dataset is a good source to use because of the many sources it has backing it up and the recency it contains as it was updated 3 months ago.
